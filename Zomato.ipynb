```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_merge=pd.merge(df,df_country,how='left')
```

```python
df_merge.head()
```

```python
df_merge.dtypes
```

```python
df_merge.columns
```

```python
df_merge.Country.value_counts()
```

```python
country_names=df_merge.Country.value_counts().index
country_values=df_merge.Country.value_counts().values
```

Pie Chart

```python
from enum import auto
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation: Most of the user are from India then united states then united kingdom

```python
rating=df_merge.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

Observation
1.When rating is between 4.5 to 4.9--->Excellent

2.When rating is between 4 to 3.5 --->very good

3.When rating is between 3.5 to 3.9--->good


```python
plt.figure(figsize=(15,6))
sns.barplot(x='Aggregate rating',y='Rating Count',data=rating,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```

```python
df_merge[ df_merge['Rating color']== 'White' ].groupby(['Country']).size().reset_index().rename(columns={0:'Rating Count'})
```

Observation
Maximum no of customers who have given zero rating are from India
